This is a classification task using marketing data from a Portugeuse bank, it aims to predict whether a client will subscribe to a term deposit given the collected data:
<ul>
<li><b>Age:</b> The age of the clients</li>
<li><b>Job:</b> A categorical feature of the job of the clients</li>
<li><b>Marital:</b>  A categorical feature of the marital status of the clients</li>
<li><b>Education:</b> The level of education of the clients</li>
<li><b>Default:</b> It tells whether a client has credit in the bank</li>
<li><b>Balance:</b> A numerical feature which specifies the actual balance in the clients' account</li>
<li><b>Housing:</b> It tells whether a client has acquired a housing loan from the bank</li>
<li><b>Loan:</b> It tells whether a client has acquired a personal loan from the bank</li>
<li><b>Contact:</b> A categorical feature which specifies the mode of communication used to contact the client</li>
<li><b>Day:</b> It tells which day of the month the client was contacted during the current campaign</li>
<li><b>Month:</b> It tells which month the client was contacted during the current campaign</li>
<li><b>Duration:</b> A numerical feature which specifies the duration of the call (in seconds) the client was contacted during the current campaign</li>
<li><b>Campaign:</b> The number of times the client was contacted during the current campaign</li>
<li><b>Pdays:</b> The number of days passed after the client was contacted in the previous campaign</li>
<li><b>Previous:</b> The number of time the client was contacted before the current campaign</li>
<li><b>Poutcome:</b> A categorical feature which specifies the outcome of the previous campaign</li>
<li><b>Deposit:</b> A categorical feature (yes or no) which specifies whether the client agreed to a term deposit during the current campaign</li>
</ul>

<h3>Exploratory Data Analysis</h3>
Here we analyse the data set to summarize the characteristics of the features using data visualization libraries (i.e. matplotlib and seaborn). EDA will help in identifying the relevant features we'll use for training the machine learning models for this use case. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Read data into a DataFrame
df = pd.read_csv("bank.csv")

# Configure parameters of the figures
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 6
plt.rcParams["figure.figsize"] = fig_size
```

```python
# Display DataFrame head
display(df.head())

# Print DataFrame dimensions
print("\nDataFrame dimensions:")
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
print(df.info())
```

```python
df.describe()
```

```python
sns.violinplot(x="deposit", y="age", data=df)
plt.xlabel("Clients subscription to a term deposit")
plt.ylabel("Age of client")
plt.show()

subs_job = df.groupby("job")["deposit"].value_counts(normalize=False).unstack()
subs_job.plot(kind="bar",stacked=True)
plt.xlabel("Client jobs")
plt.show()

subs_marital = df.groupby("marital")["deposit"].value_counts(normalize=False).unstack()
subs_marital.plot(kind="bar",stacked=True)
plt.xlabel("Client marital status")
plt.show()

subs_educ = df.groupby("education")["deposit"].value_counts(normalize=False).unstack()
subs_educ.plot(kind="bar",stacked=True)
plt.xlabel("Client education")
plt.show()
```

```python
sns.countplot(data=df,x=df['deposit'])
plt.xlabel("Clients subscription to a term deposit")
print("Number of clients who did not subscribe to a term deposit: ",df.deposit.value_counts()[0])
print("Number of clients who subscribed to a term deposit: ",df.deposit.value_counts()[1])
```

```python
sns.violinplot(x="deposit",y="balance",data=df)
plt.xlabel("Clients subscription to a term deposit")
plt.ylabel("Account balance")
plt.show()
print("Number of clients who have been contacted in a previous campaign:")
print(df.poutcome.value_counts())

# Let's determine the number of clients who have subscribed to a term deposit for this marketing campaign
# in relation to whether they were contacted during the previous campaign
subs = df.groupby("poutcome")["deposit"].value_counts(normalize=False).unstack()
subs.plot(kind="bar", stacked=False)
plt.xlabel("Previous contact status")
```

```python
# Let's examine the distribution of client contacts and days elapsed since the previous marketing campaign
sns.scatterplot(x=df.pdays,y=df.previous,size=df.previous)
plt.xlabel('Days since last contacted in previous campaign')
plt.ylabel('Prior contact frequency')
plt.legend(labels=[])
plt.show()

print("Number of clients who wasn't contacted previously:", df.pdays.value_counts()[-1])
print("Percentage of clients who were not contacted: ", round((df.pdays.value_counts()[-1]/len(df))*100,2),"%")
```

```python
# Let's examine the effectiveness of this current campaign 
sns.violinplot(x="deposit",y="campaign",data=df)
plt.xlabel("Clients subscription to a term deposit")
plt.ylabel("Contact frequency")
plt.show()
```

```python
# Let's determine  the relationship of the month and which day of the month when clients were contacted 
# to whether they subscribed to a term deposit
subs = df.groupby("month")["deposit"].value_counts(normalize=False).unstack()
subs.plot(kind="bar",stacked=True)
plt.xlabel("Contact month")
plt.show()

subs = df.groupby("day")["deposit"].value_counts(normalize=False).unstack()
subs.plot(kind="bar",stacked=True)
plt.xlabel("Contact day")
plt.show()

# Let's further investigate how the mode and duration of communication affect the clients' decision
# to whether they subscribed to a term deposit
subs = df.groupby("contact")["deposit"].value_counts(normalize=False).unstack()
subs.plot(kind="bar",stacked=True)
plt.xlabel("Mode of communication")
plt.show()

sns.violinplot(x="deposit",y="duration",data=df)
plt.xlabel("Clients subscription to a term deposit")
plt.ylabel("Duration of call in seconds")
plt.show()
```

```python
# Let's examine whether these clients have availed loans in the bank, either personal loan or housing loan
subs_loan = df.groupby("loan")["deposit"].value_counts(normalize=False).unstack()
subs_loan.plot(kind="bar",stacked=False)
plt.xlabel("Clients with personal loan")
plt.show()

subs_housing = df.groupby("housing")["deposit"].value_counts(normalize=False).unstack()
subs_housing.plot(kind="bar",stacked=False)
plt.xlabel("Clients with housing loan")
plt.show()

subs_credit = df.groupby("default")["deposit"].value_counts(normalize=False).unstack()
subs_credit.plot(kind="bar",stacked=False)
plt.xlabel("Clients with credit")
plt.show()
```

<h3>Preprocessing</h3>
In this part, a few preprocessing procedures will be implemented with the goal of improving the data quality, reducing dimensionality of data, and optimizing data for the different machine learning algorithms that will be used.

```python
# Convert binary categorical variables (yes or no) to binary encoding ( 1 or 0)
df.loan = df.loan.map({"yes":1,"no":0})
df.default = df.default.map({"yes":1,"no":0})
df.housing = df.housing.map({"yes":1,"no":0})
df.deposit = df.deposit.map({"yes":1,"no":0})
display(df.head())
```

```python
# Convert other features with categorical values to numerical values using one-hot encoding
# First, we determine the unique values of these features and assess the relevance of these values
column_list = ["job","marital","education","contact","month","poutcome"]
unique_elements = {}
for column in column_list:
    unique_elements[column] = df[column].unique()
    
print(unique_elements)

# Implement one-hot encoding 
one_hot_encoded = pd.get_dummies(df[column_list])
df.drop(columns=column_list,inplace=True)

# Remove columns associated with unknown values for the considered features
suffix = "_unknown"
column_list.remove("marital")
column_list.remove("month")

columns_to_drop = []
for element in column_list:
    columns_to_drop.append(element + suffix)
    
columns_to_drop.append("marital_divorced")
one_hot_encoded.drop(columns=columns_to_drop,inplace=True)

df = pd.concat([df,one_hot_encoded],axis=1)
display(df.info())
```
